```julia
include("OQS_Tools_v1.jl")
using Plots
using ITensors
using ITensorMPS
using LinearAlgebra
using PolyChaos
using JLD2
using .OQS_Tools_v1
BLAS.set_num_threads(4)
ITensors.enable_threaded_blocksparse(false)
```

```julia
bath_shape = "elliptical" #change me
Ds = collect(0.5:0.5:5.0)
gs = collect(0.05:0.05:0.5)
betas = [1.0,10.0,100.0,1000.0]
N = 50
mu = 0.0
dt = 0.1
Esys = 0.0
sys = N+1
sys_occ = 0.0
tmax = 5.0

times = dt:dt:tmax
```

```julia
save_dir = raw"E:\Project_Data\1Site" #Windows PC 
```

```julia
tot = length(Ds) * length(gs) * length(betas)
count = 0
for D in Ds
    for g in gs 
        for beta in betas
            J = create_spectral(bath_shape, D, g)
            J1, J2 = thermofield_transform(J, beta, mu)
            E1, h1 = chain_map(J1, N, D)
            E2, h2 = chain_map(J2, N, D)
            
            psi0, sites, states = prepare_MPS(N, sys_occ)
            H_mpo = make_H_mpo(E1, E2, h1, h2, sites, Esys, N, sys)

            mps_evolution = evolve_MPS_2(psi0, H_mpo, dt, tmax)
            filename = joinpath(save_dir, "mps_evolution_$(bath_shape)D$(D)g$(g)Esys$(Esys)beta$(beta)mu$(mu)dt$(dt).jld2")
            @save filename times mps_evolution

            count += 1
            println("$count of $tot iterations complete.")
        end
    end
end







```

```julia
for D in Ds
    for g in gs
        for beta in betas
            filename = joinpath(save_dir, "mps_evolution_$(bath_shape)D$(D)g$(g)Esys$(Esys)beta$(beta)mu$(mu)dt$(dt).jld2")
            @load filename times mps_evolution
        end
    end
end
```
