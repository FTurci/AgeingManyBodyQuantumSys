```julia
include("OQS_Tools_v1.jl")
using Plots
using LinearAlgebra
using PolyChaos
using JLD2
using .OQS_Tools_v1
BLAS.set_num_threads(4)
```

```julia
bath_shape = "flat" #change me
Ds = [1.0]
gs = [0.1,0.3,0.8]
betas = [1.0,10.0,100.0,1000.0]
mus = [-0.2,-0.1,0.0,0.1,0.2]
N = 50
dt = 0.1
Esys = 0.0
sys = N+1
sys_occ = 0.0
tmax = 50.0

sys_occs = collect(0.0:0.02:1.0)
times = 0:dt:tmax
```

```julia
save_dir = raw"E:\Project_Data\1Site" #Windows PC 
```

```julia
tot = length(Ds) * length(gs) * length(betas) * length(mus) * length(sys_occs)
count = 0 
for D in Ds
    for g in gs
        for beta in betas
            for mu in mus
                J = create_spectral(bath_shape, D, g)
                J1, J2 = thermofield_transform(J, beta, mu)
                E1, h1 = chain_map(J1, N, D)
                E2, h2 = chain_map(J2, N, D)
                H = make_H_matrix(E1, E2, h1, h2, Esys, N, sys)
                data = Any[]
                for sys_occ in sys_occs
                    C0 = prepare_correlations(N, sys, sys_occ)
                    Cs = evolve_correlations(C0, H, times, N)
                    nSys = real.(Cs[sys,sys,:])
                    push!(data, copy(nSys))
                    count += 1
                    println("$count of $tot iterations complete.")
                end
                filename = joinpath(save_dir, "exact_occs_$(bath_shape)D$(D)g$(g)Esys$(Esys)beta$(beta)mu$(mu)dt$(dt).jld2")
                @save filename times data
            end
        end
    end
end
```
